## VISUALISATION

```python
sns.pairplot(iris,hue='species',height=3,aspect=1);
```

```python
iris.hist(edgecolor='black',linewidth=1.2,figsize=(12,8));
plt.show();
```

```python
plt.figure(figsize=(12,8));
plt.subplot(2,2,1)
sns.violinplot(x='species',y='sepal_length',hue='species',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='sepal_width',hue='species',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='petal_length',hue='species',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='petal_width',hue='species',data=iris)
```

```python
iris.boxplot(by='species',figsize=(12,8));
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris,figsize=(12,10))
plt.show()
```
